### Data Preparation and Cleaning

 Once we have the data in front of us we will rename the columns, check the number of rows and columns in the dataset, check the data type of each attribute, find null values, remove unnecessary rows and columns, conver the data type and check outliers.

```python
#Adding the column name
df_new = df.rename(columns={"Unnamed: 0":"Date"})
df_new
```

```python
#shape of the dataset
df_new.shape
```

The shape attribute tells us about the dimension of the dataset. Our dataset has 329 rows and 49 columns.

```python
df_new.info()
```

info() is used to verify the data and datatype of each attribute. By using info(), we discovered that 3 columns named 'Unnamed: 34', 'Unnamed: 37' and 'Unnamed: 47' have zero non-null values. Hence, we will drop these columns. Also the first row of the dataset contains postcodes of the areas which is of no use to us. Therefore, we will drop the first row as well.

Additionally, we noticed that the data type of the rest of the columns is object. Hence we will convert it to float to make our further analysis easier.

```python
#Dropping unwanted row and columns
df_new = df_new.drop(df_new.columns[[34,37,47]], axis = 1).drop(df_new.index[[0]])
df_new
```

```python
#Converting the data type from object to float
df_new.iloc[:,1:] = df_new.iloc[:,1:].astype(float)
```

```python
df_new.info()
```

From the above function, we have converted all the columns into float and we discovered that we don't have any null values in our dataset.

```python
#Creating a separate list for boroughs
inner_london = ['Camden', 'Greenwich', 'Hackney', 'Hammersmith & Fulham', 'Islington', 'Kensington & Chelsea', 'Lambeth', 'Lewisham', 'Southwark', 'Tower Hamlets', 'Wandsworth', 'Westminster', 'City of London']
outer_london = ['Barking & Dagenham', 'Barnet', 'Bexley', 'Brent', 'Bromley', 'Croydon', 'Ealing', 'Enfield', 'Haringey', 'Harrow', 'Havering', 'Hillingdon', 'Hounslow', 'Kingston upon Thames', 'Merton', 'Newham', 'Redbridge', 'Richmond upon Thames', 'Sutton', 'Waltham Forest']
```

```python
#Checking for outliers for the inner London boroughs
fig, ax = plt.subplots(figsize=(15, 8))
df_new.boxplot(column=inner_london, ax=ax)
plt.setp(ax.get_xticklabels(), rotation=90);
```

```python
#Checking for outliers for the outer London boroughs
fig, ax = plt.subplots(figsize=(15, 8))
df_new.boxplot(column=outer_london, ax=ax)
plt.setp(ax.get_xticklabels(), rotation=90, ha="right");
```

```python
combined_boroughs = inner_london.copy()
```

```python
combined_boroughs.extend(outer_london)
```

```python
#Created a new list of inner London and outer London boroughs combined
combined_boroughs
```

```python
combined_boroughs.append('Inner London')
combined_boroughs.append('Outer London')
combined_boroughs.append('LONDON')
```

```python
# getting all the remaining column for boxplot visualisation
remaining_columns = [column for column in df_new.columns if column not in combined_boroughs and column != 'Date']
remaining_columns
```

```python
#Checking for outliers for the rest of regions of England
fig, ax = plt.subplots(figsize=(15, 8))
df_new.boxplot(column=remaining_columns, ax=ax)
plt.setp(ax.get_xticklabels(), rotation=90, ha="right");
```

Outliers are data points in a dataset that are far apart from the rest of the data. In other words, they are isolated values in a dataset. Outliers may pose a problem in the analysis because they may cause us to miss key findings or manipulate real results. Boxplots are commonly used to identify outliers. If there are any outliers, the boxplot will display some of the data points that fall outside the normal range of the data.
From above method, we discovered that our dataset is free from outliers.

### Exploratory Data Analysis and Data Visualisation

In this section, we will perform descriptive statistics to better understand the dataset's features, explore the distribution of average house prices based on borough, find out minimum and maximum house prices over the years using scatter plot, compare the average price in inner London and outer London and will see the yearly housing price trend in the UK.

```python
stats = df_new[df_new .columns[1:]]
stats.describe()
```

We can see some of the dataset's most important characteristics using the describe() method. It demonstrates the spread of data for numeric value. The minimum and maximum value, mean value, various percentile values, are all clearly visible.

We will now compare the average house price of inner and outer London over the years. 

```python
#Comparing the average house prices in inner and outer London from 1995 to 2022
fig, ax = plt.subplots(figsize=(15, 8))
ax.plot(df_new['Date'], df_new['Inner London'], color='blue', label='Inner London')
ax.plot(df_new['Date'], df_new['Outer London'], color='red', label='Outer London')
plt.legend()
plt.title('Average House Prices in Inner and Outer London')
plt.xlabel('Year')
plt.ylabel('Average Prices')
plt.show()
```

Insight: Our line graph shows that prices in inner London are significantly higher than prices in outer London. In 1995, there was little difference in average house prices between inner and outer London. However, the price difference continued to grow over time. There was also a decline in 2009, but once that period was over, prices continued to rise rapidly.
